```python
import pandas as pd

# Load the dataset
file_path = "Manufacturing_Dataset.csv"  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Analyze missing values
missing_values = df.isnull().sum()
print("missing_values")
print(missing_values)
```

```python
print(" len(df)")
print(len(df))
```

```python
missing_percentage = (missing_values / len(df)) * 100
print(f"Missing Percentage:\n{missing_percentage}")
```

```python
pd.DataFrame({"Column": df.columns,
    "Missing Values": missing_values,
    "Percentage Missing (%)": missing_percentage
}).sort_values(by="Percentage Missing (%)",  ascending=False)
```

```python
# Handle missing values
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ['float64', 'int64']:  # Numerical column
            df[col].fillna(df[col].mean(), inplace=True)
        else:  # Categorical column
            df[col].fillna(df[col].mode()[0], inplace=True)

# Save the imputed dataset
df.to_csv("imputed_data.csv", index=False)

print("\nMissing values imputed and saved to 'imputed_data.csv'.")
```
