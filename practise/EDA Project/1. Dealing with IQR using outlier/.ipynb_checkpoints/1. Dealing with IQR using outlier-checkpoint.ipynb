```python
import pandas as pd

# importing data
data = pd.read_csv('Manufacturing_Dataset.csv')

df = pd.read_csv(data)
```

```python
# Function to calculate IQR and detect outliers
def calculate_iqr_and_outliers(df):
    numerical_columns = df.select_dtypes(include='number').columns
    outliers_info = {}

    for col in numerical_columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Detect outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outliers_info[col] = {
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'Lower Bound': lower_bound,
            'Upper Bound': upper_bound,
            'Outliers': outliers[col].tolist()
        }

    return outliers_info

# Calculate IQR and outliers
outliers_result = calculate_iqr_and_outliers(df)

# Display results
for column, info in outliers_result.items():
    print(f"Column: {column}")
    for key, value in info.items():
        print(f"  {key}: {value}")
    print("\n")



```
