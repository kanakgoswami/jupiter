```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Load the dataset
file_path = "Manufacturing_Dataset.csv"  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Columns of interest
shift_col = "Shift"  # Replace with your shift column name
time_col = "Production Time Hours"  # Replace with your production time column name
energy_col = "Energy Consumption kWh"  # Replace with your energy consumption column name

# Check if columns exist
if all(col in df.columns for col in [shift_col, time_col, energy_col]):
    # Group by shifts and calculate averages
    shift_summary = df.groupby(shift_col)[[time_col, energy_col]].mean()
    print("Average Production Time and Energy Consumption by Shift:")
    print(shift_summary)

    # Bar plots for visualization
    shift_summary.plot(kind="bar", figsize=(10, 6), grid=True)
    plt.title("Average Production Time and Energy Consumption by Shift")
    plt.xlabel("Shift")
    plt.ylabel("Average Values")
    plt.legend(["Average Production Time", "Average Energy Consumption"])
    plt.xticks(rotation=0)
    plt.show()

    # Statistical tests (ANOVA) to check for significant differences
    day_data = df[df[shift_col] == "Day"]
    swing_data = df[df[shift_col] == "Swing"]
    night_data = df[df[shift_col] == "Night"]

    # Production time comparison
    f_stat_time, p_val_time = f_oneway(day_data[time_col], swing_data[time_col], night_data[time_col])
    print(f"ANOVA for Production Time: F-statistic = {f_stat_time:.2f}, p-value = {p_val_time:.4f}")

    # Energy consumption comparison
    f_stat_energy, p_val_energy = f_oneway(day_data[energy_col], swing_data[energy_col], night_data[energy_col])
    print(f"ANOVA for Energy Consumption: F-statistic = {f_stat_energy:.2f}, p-value = {p_val_energy:.4f}")

    # Interpretation
    if p_val_time < 0.05:
        print("Significant differences exist in production time across shifts.")
    else:
        print("No significant differences in production time across shifts.")

    if p_val_energy < 0.05:
        print("Significant differences exist in energy consumption across shifts.")
    else:
        print("No significant differences in energy consumption across shifts.")
else:
    print("Error: One or more columns are not found in the dataset.")
```
