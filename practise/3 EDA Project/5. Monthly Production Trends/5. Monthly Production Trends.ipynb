```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "Manufacturing_Dataset.csv"  # Replace with your file path
df = pd.read_csv(file_path)

# Extract month from the date column (replace 'Production_Date' with your date column name)
df['Production_Date'] = pd.to_datetime(df['Date'])  # Ensure date is in datetime format
df['Month'] = df['Production_Date'].dt.month

# Group by month and calculate average units produced
monthly_trends = df.groupby('Month')['Units Produced'].mean()
print("Average Units Produced per Month:")
print(monthly_trends)

# Plotting the trends
plt.figure(figsize=(10, 6))
monthly_trends.plot(kind='line', marker='o')
plt.title("Monthly Production Trends")
plt.xlabel("Month")
plt.ylabel("Average Units Produced")
plt.grid(True)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
