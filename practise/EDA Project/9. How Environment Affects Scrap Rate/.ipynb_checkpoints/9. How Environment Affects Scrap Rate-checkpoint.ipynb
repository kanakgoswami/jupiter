```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "Manufacturing_Dataset.csv"  # Replace with your file path
df = pd.read_csv(file_path)

# Ensure the required columns exist
temperature_col = "Average Temperature C"    # Replace with the column representing temperature
humidity_col = "Average Humidity Percent"          # Replace with the column representing humidity
scrap_rate_col = "Scrap Rate"      # Replace with the column representing scrap rate

# Ensure the columns are present in the dataset
if all(col in df.columns for col in [temperature_col, humidity_col, scrap_rate_col]):
    # Calculate correlations
    correlation_temp_scrap = df[temperature_col].corr(df[scrap_rate_col])
    correlation_humidity_scrap = df[humidity_col].corr(df[scrap_rate_col])
    correlation_temp_humidity = df[temperature_col].corr(df[humidity_col])

    print(f"Correlation between Temperature and Scrap Rate: {correlation_temp_scrap:.2f}")
    print(f"Correlation between Humidity and Scrap Rate: {correlation_humidity_scrap:.2f}")
    print(f"Correlation between Temperature and Humidity: {correlation_temp_humidity:.2f}")
    
    # Visualizing correlations with scatter plots
    plt.figure(figsize=(12, 5))
    
    # Scatter plot for temperature vs. scrap rate
    plt.subplot(1, 2, 1)
    plt.scatter(df[temperature_col], df[scrap_rate_col], alpha=0.6, color='red')
    plt.title("Temperature vs. Scrap Rate")
    plt.xlabel("Temperature")
    plt.ylabel("Scrap Rate")

    # Scatter plot for humidity vs. scrap rate
    plt.subplot(1, 2, 2)
    plt.scatter(df[humidity_col], df[scrap_rate_col], alpha=0.6, color='blue')
    plt.title("Humidity vs. Scrap Rate")
    plt.xlabel("Humidity")
    plt.ylabel("Scrap Rate")

    plt.tight_layout()
    plt.show()

    # Heatmap of correlation matrix for better visualization
    correlation_matrix = df[[temperature_col, humidity_col, scrap_rate_col]].corr()
    plt.figure(figsize=(6, 5))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1)
    plt.title("Correlation Heatmap")
    plt.show()

else:
    print("Error: One or more required columns are not in the dataset.")
```
