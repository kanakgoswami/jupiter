```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "Manufacturing_Dataset.csv"  # Replace with your CSV file path
df = pd.read_csv(file_path)
```

```python
# Check if the columns exist
if material_cost_col in df.columns and labor_cost_col in df.columns:
    # Scatter plot to visualize the relationship
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[material_cost_col], y=df[labor_cost_col])
    plt.title("Relationship Between Material Cost Per Unit and Hourly Labor Cost")
    plt.xlabel("Material Cost Per Unit")
    plt.ylabel("Hourly Labor Cost")
    plt.grid(True)
    plt.show()

    # Calculate correlation
    correlation = df[[material_cost_col, labor_cost_col]].corr(method="pearson")
    print("Correlation Matrix:")
    print(correlation)

    # Interpretation
    if abs(correlation.loc[material_cost_col, labor_cost_col]) > 0.7:
        print("Strong correlation detected.")
    elif abs(correlation.loc[material_cost_col, labor_cost_col]) > 0.3:
        print("Moderate correlation detected.")
    else:
        print("Weak or no correlation detected.")
else:
    print("Error: Columns not found in the dataset.")
```
